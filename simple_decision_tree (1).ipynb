# Decision Tree

```python
# import the necessary modules

import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

```python
# read dataframe

df = pandas.read_csv('c:/Users/Usuario/Documents/7mo Semestre/Minería de datos/lab.csv')
print(df)
```

```python
# change string values into numerical

d1 = {'ECUATORIANO': 0, 'COLOMBIANO': 1, 'VENEZOLANO': 2}
df['Nacionalidad'] = df['Nacionalidad'].map(d1)

d2 = {'YES': 1, 'NO': 0}
df['Seguro'] = df['Seguro'].map(d2)

print(df)
```

```python
# choose features, i.e., columns from which we try to predict
features = ['Edad', 'ExperienciaLaboral', 'Calificacion', 'Nacionalidad']
X = df[features]

# choose target, i.e., the column with the values we try to predict
y = df['Seguro']

print(X, y, sep='\n\n')
```

```python
# create a decision tree

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
```

```python
# save decision tree as an image

data = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('mydecisiontree.png')
```

```python
# show the image

img = pltimg.imread('mydecisiontree.png')
imgplot = plt.imshow(img)
plt.show()
```
